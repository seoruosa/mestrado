```python
import os 

folderpath = 'src/cpp/build/'

results_map = {}
for file in os.listdir(folderpath):
    filepath = os.path.join(folderpath, file)
    list = []
    
    if os.path.isfile(filepath) and ('nsgaii' in file):
        with open(filepath, mode='r') as f:
            reader = csv.reader(f, delimiter=';')
            i = 0
            for line in reader:
                if i > 0:
                    list.append([float(el.strip().replace(',', '.')) for el in line if el != ' '][-2:])
                i = i + 1
        results_map[file] = list

all_results = {key:results_map[key] for key in results_map if len(results_map[key]) > 0}
all_results.keys()
```

```python
all_pop = []

for key in all_results:
    all_pop.extend(all_results[key])
#     print(len(all_results[key]))

len(all_pop)
```

```python
import numpy as np

hv = pg.hypervolume(all_pop)

# hv.compute()

ref_point = hv.refpoint(offset=10)
ref_point

print(f"FILE\t\t RESULT")
for key in all_results:
    hv = pg.hypervolume(all_results[key])
    s_metric = hv.compute(ref_point)
    
    print(f"{key}\t\t{s_metric}")
```
